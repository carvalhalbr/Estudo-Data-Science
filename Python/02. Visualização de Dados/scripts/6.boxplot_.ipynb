# Boxplot

```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('E:\\DATA_SCIENCE\\Data-Science\\Python\\02. Visualização de Dados\\src\\dados\\trees.csv')
base.head()
```

```python
# Geração do boxplot
# patch_artist = True preenche, showfliers outliers
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True,patch_artist = True)
# vert = define se vai ser horizontal ou vertical; showfliers = define se terá outliers ou não; notch = cria ou não um entalhe no centro; patch_artist = preenche o boxplot
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
#dados por linha
plt.boxplot(base)
plt.title('Árvores')
plt.xlabel('Dados')
# por algum motivo o VS Code não está compilando essa função corretamente, porém é compilada normalmente no Jupyter notar que estamos usando a base de dados toda.
```

```python
# Geração de 3 boxplots, cada um mostrado informações diferentes
plt.boxplot(base.Volume, vert = False)
plt.boxplot(base.Girth, vert = False)
plt.boxplot(base.Height, vert = False)
plt.title('Árvores')
plt.xlabel('Dados')
```
